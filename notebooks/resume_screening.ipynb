```python
job_skills = [
    "python",
    "sql",
    "java",
    "machine learning",
    "pandas",
    "numpy",
    "statistics",
    "excel"
]

scores = []

for index, row in df.iterrows():
    resume_text = str(row['Resume_str']).lower()

    score = 0

    for skill in job_skills:
        if skill in resume_text:
            score += 1

    percentage = (score / len(job_skills)) * 100

    scores.append(percentage)

df['Score'] = scores
```

```python
def get_matched_skills(resume_text, job_skills):
    matched = []

    for skill in job_skills:
        if skill in resume_text.lower():
            matched.append(skill)

    return matched
```

```python
df['Matched_Skills'] = df['Resume_str'].apply(
    lambda x: get_matched_skills(str(x), job_skills)
)

df[['Category', 'Score', 'Matched_Skills']].head()
```

```python
def get_missing_skills(resume_text, job_skills):
    missing = []

    for skill in job_skills:
        if skill not in resume_text.lower():
            missing.append(skill)

    return missing
```

```python
df['Missing_Skills'] = df['Resume_str'].apply(
    lambda x: get_missing_skills(str(x), job_skills)
)

df[['Category', 'Score', 'Matched_Skills', 'Missing_Skills']].head()
```

```python
top_candidates = df.sort_values(
    by='Score',
    ascending=False
)

top_candidates.to_csv(
    '../output/top_candidates.csv',
    index=False
)

print("Top candidates saved successfully!")
```

```python
job_description = """
Data Scientist

Skills Required:
Python
Machine Learning
SQL
Pandas
NumPy
Statistics
"""

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
resume_text = df['Resume_str'][0]

documents = [job_description, resume_text]

tfidf = TfidfVectorizer()

matrix = tfidf.fit_transform(documents)

similarity = cosine_similarity(matrix[0:1], matrix[1:2])

print("Similarity Score:", similarity[0][0] * 100)
```

```python
scores = []

for resume in df['Resume_str']:

    documents = [job_description, str(resume)]

    tfidf = TfidfVectorizer()

    matrix = tfidf.fit_transform(documents)

    similarity = cosine_similarity(
        matrix[0:1],
        matrix[1:2]
    )

    scores.append(similarity[0][0] * 100)

df['Similarity_Score'] = scores
```

```python
top_candidates = df.sort_values(
    by='Similarity_Score',
    ascending=False
)

top_candidates[
    ['Category', 'Similarity_Score']
].head(10)
```
